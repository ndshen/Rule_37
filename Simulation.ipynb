```python
#the code of class simulator is in another python file under the same directory
from Simulate import simulator
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy.stats import norm

norm_rv = norm()
norm_sim = simulator(rv=norm_rv, sample_size=100, simulate_freq=100000, m_scale=100, sample_label='norm sample')
norm_sim.run(rv_plot=True)
```

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,10))
x = np.linspace(uniform.ppf(0.01), uniform.ppf(0.99), 100)
uni_rv = uniform()
ax.plot(x, uni_rv.pdf(x), lw=4, label='frozen pdf', color='#fdc086')
ax.set_title('pdf of the random variable')

uni_sim = simulator(rv=uni_rv, sample_size=100, simulate_freq=100000, m_scale=100, sample_label='uni sample')
uni_sim.run()
```

```python
from scipy.stats import skewnorm

fig, ax = plt.subplots(figsize=(15,10))
a = -4
x = np.linspace(skewnorm.ppf(0.01, a), skewnorm.ppf(0.99, a), 100)
left_rv = skewnorm(a)
ax.plot(x, left_rv.pdf(x), lw=4, label='frozen pdf', color='#fdc086')
ax.set_title('pdf of the random variable')

left_skew_sim = simulator(rv=left_rv, sample_size=100, simulate_freq=100000, m_scale=100, sample_label='left skew sample')
left_skew_sim.run()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
a = 4
x = np.linspace(skewnorm.ppf(0.01, a), skewnorm.ppf(0.99, a), 100)
right_rv = skewnorm(a)
ax.plot(x, right_rv.pdf(x), lw=4, label='frozen pdf', color='#fdc086')
ax.set_title('pdf of the random variable')

right_skew_sim = simulator(rv=right_rv, sample_size=100, simulate_freq=100000, m_scale=100, sample_label='right skew sample')
right_skew_sim.run()
```
